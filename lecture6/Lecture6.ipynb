# Higher Dimensional plots
## Quiver plot 
Plots vector fields!

````python
#some data
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z, .2, .2)

plt.quiver(x,y,u,v)
plt.show()
````

## Contour plot
3-D plot where the height represents output value and the x-y plane is the input

````python
# some data
feature_x = np.arange(0, 100, 2)
feature_y = np.arange(0, 100, 3)
[X, Y] = np.meshgrid(feature_x, feature_y)
Z = np.cos(X / 2) + np.sin(Y / 4)

plt.contour(X,Y,Z)
plt.show()
````

## Heat Maps
Similar to contour plots, but the height is represented by colours.

````python
plt.imshow(Z, cmap='hot') # there are a lot of cmap values check the docs for more
plt.show()
````

## Array Interpolation schemes

Interpolation calculates what the color or value of a pixel "should" be, according to different mathematical schemes. One common place that this happens is when you resize an image. The number of pixels change, but you want the same information. Since pixels are discrete, there's missing space. Interpolation is how you fill that space.


This is why your images sometimes come out looking pixelated when you blow them up. The effect is more pronounced when the difference between the original image and the expanded image is greater. Let's take our image and shrink it. We're effectively discarding pixels, only keeping a select few. Now when we plot it, that data gets blown up to the size on your screen. The old pixels aren't there anymore, and the computer has to draw in pixels to fill that space.

````python
# We use the PIL library to resize and load the image
from PIL import Image

img = Image.open("C:/Users/91788/Downloads/matplotlib_sample_img.jpg")
img.thumbnail((64, 43), Image.ANTIALIAS) # Resizes the image to 64 x 43 [Preserving the aspect ratio]
````

````python
print(img.size)
````

````python
small_imgplot = plt.imshow(img)
````

````python
nearest_interpolation_img = plt.imshow(img, interpolation="nearest")
````

````python
bicubic_interpolation_img = plt.imshow(img, interpolation="bicubic")
````

## Subplots

`fig.add_subplot(rows, cols, num)` allows us to create a grid of plot panels that can show any sort of plot. (Note that num = number of subplots  counting from left to right, top to bottom and indexed starting at 1). 

````python
fig = plt.figure()
for i in range(6):
  ax = fig.add_subplot(2, 3, i+1) # i+1 because of 1-based indexing
  ax.set_title(f"Plot {i}")
````

To remove the overlap use the `subplots_adjust()` command. We'll adjust the width between each plot (wspace), and the height between each plot (hspace) as a fraction of the size of each plot:

````python
fig.subplots_adjust(wspace=0.7, hspace=0.7)
fig
````

Now we can plot whatever we want

````python
x = np.linspace(0, 5, 500)
for i in range(6):
  fig.axes[i].plot(x, x**i) # the fig has 6 subplots and we can access them this way
  fig.axes[i].set_ylabel(f'$x^{i}$')
  fig.axes[i].set_xlabel(f'$x$')
fig
````

There is another way to deal with subplots! This is the `plt.subplots()` method.

````python
fig, ax = plt.subplots() # Creates a single figure and axes
ax.plot(x, np.exp(x))
````

````python
# Now let's create multiple subplots 
fig, ax = plt.subplots(2, 3) # Creates 2x3 grid
fig.subplots_adjust(wspace=0.7, hspace=0.7)
for i in range(2):
  for j in range(3):
    ax[i, j].plot(x, (i+j)*np.sin((2*i+j)*x))
    ax[i,j].set_title(f"${i+j}\sin {2*i+j}x$")
````

In this case, we know that all the plots have the same x range and the same y range, and it would be nice if we could take away all the unnecessary labels. This can be done by specifying sharex or sharey:

````python
fig, ax = plt.subplots(2, 3, sharex=True, sharey=True) # Creates 2x3 grid
fig.subplots_adjust(wspace=0.7, hspace=0.7)
for i in range(2):
  for j in range(3):
    ax[i, j].plot(x, (i+j)*np.sin((2*i+j)*x))
    ax[i,j].set_title(f"${i+j}\sin {2*i+j}x$")
````

We can make more complicated grids using `plt.GridSpec()`

````python
fig = plt.figure(figsize=(8, 8))
gs = plt.GridSpec(3, 3)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, :2])
ax3 = fig.add_subplot(gs[1:, 2])
ax4 = fig.add_subplot(gs[2, 0])
ax5 = fig.add_subplot(gs[2, 1])
````
